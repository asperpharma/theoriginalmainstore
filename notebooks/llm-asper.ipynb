```python
import os
import google.generativeai as genai

# Set your Gemini API key (from https://aistudio.google.com/apikey or .env)
API_KEY = os.environ.get("GOOGLE_API_KEY") or os.environ.get("GEMINI_API_KEY") or "YOUR-API-KEY"
if API_KEY and API_KEY != "YOUR-API-KEY":
    genai.configure(api_key=API_KEY)
    print("Gemini configured.")
else:
    print("Set GOOGLE_API_KEY or paste your key in the cell above.")
```

```python
# Simple LLM call – change the prompt for your use case
model = genai.GenerativeModel("gemini-2.0-flash")

prompt = """You are a copywriter for Asper Beauty Shop (luxury skincare & beauty).
Write a short, elegant product description (2–3 sentences) for: Luxury hydrating face cream.
Tone: premium, minimal, trustworthy."""

response = model.generate_content(prompt)
print(response.text)
```

## Optional: Use with your catalog CSV

Load a product CSV and use the LLM to generate or refine descriptions for each
row.

```python
import pandas as pd

# Point to your catalog CSV (change path if needed)
CATALOG_PATH = "../Asper_Catalog_FINAL_READY.csv"  # or full path

try:
    df = pd.read_csv(CATALOG_PATH, nrows=5)  # first 5 rows as sample
    display(df.head())
except Exception as e:
    print("Could not load CSV:", e)
    print("Update CATALOG_PATH above.")
```

```python
# Example: generate one product description from a row
def describe_product(model, name: str, category: str) -> str:
    prompt = f"""Asper Beauty Shop copywriter. Write 2–3 sentence product description for:
Product: {name}
Category: {category}
Tone: luxury, minimal, trustworthy."""
    r = model.generate_content(prompt)
    return r.text.strip()

# Pick column names that match your CSV (e.g. Title, Product Type, etc.)
if not df.empty and "Title" in df.columns:
    name = df.iloc[0].get("Title", "Unknown product")
    cat = df.iloc[0].get("Product Type", df.iloc[0].get("Category", "Beauty"))
    desc = describe_product(model, name, cat)
    print(desc)
else:
    print("No 'Title' column or empty DF. Adjust column names in the code.")
```
